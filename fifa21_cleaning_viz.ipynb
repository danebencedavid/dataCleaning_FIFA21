```python
# Not all columns are needed, some columns are not the right type.
# Columns we can get rid of:
# Loan Date End,Release Clause,Photo URL, Player URL

df.drop(['Loan Date End','Release Clause','photoUrl','playerUrl'],axis=1,inplace=True)
df.head()
```

```python
df.shape
```

```python
# Replaceing LongName with Name
df.rename(columns={'LongName':'Name'},inplace=True)
```

```python
df.head()
```

```python
# Getting rid of \n in Team & Contract and splitting it to 2 columns
df['Team & Contract'] = df['Team & Contract'].astype('str')
df['Team & Contract'].replace('\n','',inplace=True,regex=True)
```

```python
team = []
for x in range(len(df['Team & Contract'])):
    text = df['Team & Contract'][x]
    index = text.find("~")
    if index != -1:
        team_name = text[:index-5].strip()
    team.append(team_name)
```

```python
contract = []
for x in range(len(df['Team & Contract'])):
    text = df['Team & Contract'][x]
    index = text.find("~")
    if index != -1:
        c = text[index-5:index+6].strip()
    contract.append(c)
```

```python
len(team)
```

```python
len(contract)
```

```python
df = df.drop(columns = ['Team & Contract'])
df['Team'] = team
df['Contract Duration'] = contract
```

```python
hits = []
for x in range(len(df['Hits'])):
    y = str(df["Hits"][x])
    y = y.replace('\n','')
    y = y.replace('n','')
    y = y.replace('K','00')
    y = y.replace('.','')
    hits.append(y)
df['Hits'] = hits
df = df.astype({'Hits':'int'})
```

```python
df['Height'].unique()
```

```python
def ftcm(x):
    if x[-1] == '"':
        x = x.replace("\"","")
        inch = int(x[2:]) * 2.54
        foot = int(x[0]) * 30.48
        return round(foot+inch)
```

```python
df['Height'] = df['Height'].apply(ftcm)
```

```python
df['Height'].unique()
```

```python
df.head()
```

```python
#Removing the star (★) in IR, W/F,SM columns.

def no_star(x):
    x = x.replace('★','')
    return x
```

```python
df['IR'] = df['IR'].apply(no_star)
```

```python
df['W/F'] = df['W/F'].apply(no_star)
df['SM'] = df['SM'].apply(no_star)
```

```python
df.head()
```

```python
#Converting 'Joined' column's type from object to datetime.
df['Joined'] = pd.to_datetime(df['Joined'])
```

```python
df.head()
```

```python
def no(x):
    x = x.replace('lbs','')
    return x
```

```python
df['Weight'] = df['Weight'].apply(no)
```

```python
def no_1(x):
    x = x.replace('€','')
    x = x.replace('K','')
    return x
```

```python
df['Wage'] = df['Wage'].apply(no_1)
```

```python
df['Wage'] = df['Wage'].astype('int')
```
